```python
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries imported successfully")
```

```python
dataset = pd.read_csv("cleaned_students.csv")

print("Dataset loaded successfully")
dataset
```

```python
plt.figure(figsize=(10, 6))

plt.bar(dataset["Name"], dataset["Marks"])

plt.title("Marks Scored by Students")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv("cleaned_students.csv")

print("Dataset:")
print(dataset)

# 1. Bar chart
plt.figure(figsize=(10, 6))
plt.bar(dataset["Name"], dataset["Marks"])
plt.title("Marks Scored by Students")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.tight_layout()
plt.savefig("student_marks_bar_chart.png")
plt.show()

# 2. Line chart
plt.figure(figsize=(10, 6))
plt.plot(
    dataset["Name"],
    dataset["Marks"],
    marker="o"
)
plt.title("Student Marks Trend")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.grid(True)
plt.tight_layout()
plt.savefig("student_marks_line_chart.png")
plt.show()

# 3. Pie chart
course_count = dataset["Course"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    course_count.values,
    labels=course_count.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Distribution of Students by Course")
plt.axis("equal")
plt.savefig("course_distribution_pie_chart.png")
plt.show()

print("All charts created and saved successfully.")
```
